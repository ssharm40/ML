## Linear, Ridge, Lesso

```python
# Housing price dataset

from sklearn.datasets import load_boston

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= load_boston()
#df
```

```python
type(df)
```

```python
dataset= pd.DataFrame(df.data)
dataset.columns= df.feature_names


dataset.head()
```

```python
dataset['Price']=df.target
```

```python
dataset.head()
```

```python
#Dividing the dataset into independent and depedent features

X=dataset.drop('Price', axis=1)##independent features
Y=dataset['Price']##dependent features

```

```python
X.head()
```

```python
Y.head()
```

```python
# Let's start with ml
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(X,Y,test_size=0.33, random_state=42)
```

```python

from sklearn.linear_model import LinearRegression

from sklearn.model_selection import cross_val_score
#It will go to apply multiple train test



lin_reg= LinearRegression()


# Here we apply the Cross Validation---- In cross validation we divide our train and test data in such a way that every 
# combination of train and test data is taken by the model and whose ever accuracy is better take that.

#Basically first we do the train test then after that we will apply that
mse= cross_val_score(lin_reg, x_train,y_train, scoring='neg_mean_squared_error',cv=5)

mean_mse= np.mean(mse)
print(mean_mse)
```

```python
# Let's apply Ridge

from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
#Gridsearch cv----- help you in hyper-parameter turing


ridge=Ridge()


#define parameter
params={'alpha':[1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1,5,10,20]}

#So it will aply for each and then selected anyone which one giving you the best accuracy

ridge_regressor= GridSearchCV(ridge, params, scoring='neg_mean_squared_error', cv=15)

ridge_regressor.fit(x_train,y_train)
```

```python
# Let's select the best parameter
print(ridge_regressor.best_params_, ridge_regressor.best_score_)
```

```python
# Let's apply Lasso

from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV


lasso= Lasso()

param={'alpha':[1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1,5,10,20]}

lasso_regressor= GridSearchCV(lasso, params, scoring='neg_mean_squared_error', cv=15 )
lasso_regressor.fit(x_train,y_train)
```

```python
# Let's select the best parameter
print(lasso_regressor.best_params_, lasso_regressor.best_score_)
```

```python
y_pred= lasso_regressor.predict(x_test)

from sklearn.metrics import r2_score

score= r2_score(y_pred, y_test)

print(score)
```

```python
from sklearn.linear_model import LogisticRegression

# In logistic we have l1,l2 norms also


params=[{'C':[1,5,10]}, {'max_iter': [100,150]}]

model1= LogisticRegression(C=100, max_iter=100)

model= GridSearchCV(model1, param_grid=params, scoring='f1', cv=5)
```

```python
# Let's apply confusion matrix
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score

confusion_matrix(y_pred, y_test)
```

```python
accuracy_score(y_pred, y_test)
```

##                                                          Naive Bayes Algorith


It's a classification algorithm.

Works on probablity Bayes Theorem

### What is this Bayes Theorem?

In prob we have two type of event------ 1) Independent Even                   2) Dependent Event


In Bayes theorem we will work on dependent event.

Exp- Suppose we have a bag with **2 Red** and **3 White Ball** so now for the first time when you pick one red ball then the probablity of picking red ball will be - P(Red)= 2/5
And In second time probablity of picking Red Ball will be------ P(Red)= 1/4

So now second probablity is depend on first probablity then these type of event is known as **Dependent Event**



P(Red and White)= P(Red) * P(White/Red)
And this formula is known as **Conditional Probablity**


So as per the rulewe can say: 

P(A and B)           =           P(B and A) 
P(A) * P(B/A)        =           P(B) * P(A/B)



P (B/A)              =           [P(B) * P(A/B] / P(A)



And this is know as **Bayes Theorem**





As in ML Dataset we have multiple feature so this theorem will be look like this: 

P(Yes/ X1,X2....Xn)= [P(Yes) * P(X1/Yes) * P(X2/Yes)-----------P(Xn/Yes)] **/** [P(x1) * P(x2)-----P(Xn)]


P(No/ X1,X2.....Xn)= [P(No) * P(X1/No) * P(X2/No)-----------P(Xn/No)] **/** [P(x1) * P(x2)-----P(Xn)]



So as we see in both the above function Dominator is same so we can neglect/ignore the Dominator

After getting the result we have to normalize the answer


P(Yes/Xi)= [P(Yes)] / [P(Yes) + P(No)]

P(No/Xi)= [P(No)] / [P(Yes) + P(No)]       OR    1- P(Yes/Xi)

#                                       KNN(K Nearest Neighbour)

==> works on both - Classification and Regression

As the name say K Nearest Neighbour mean we need to find the K(any positive integer) nearby to the point that we get.

So in classification we will majority of neighbour belong to which class that will be our answer
   In Regression we will take mean of nearest neighbour and that will be our answer.
   
   

### How do we find the nearest neighbour:

Two technique we use-------
1) Equidistant Distance

2) Manhattan Distnace

### So how you will select the K Value:
K Value is a Hyperparameter variable
So we will play with the multiple K value by using loop(exp 1 to 30 or anything else) and then check the mse(Mean square error)
for each K value, and whatever value gives the best result will be our answer.
